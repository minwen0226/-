```python
#練習題2
for num in range (1, 100):
    if not num%3 and not num%5:
        print ("⼗五的倍數")
    elif not num%3:
        print ("三的倍數")
    elif not num%5:
        print ("五的倍數")
    else:
        print(num)
```
